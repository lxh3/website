Hyperbolic Fixed Points
---------------------------

```python
def E(x, y, lx, ly):
    """Return the flow vector for a hyperbolic fixed point with eigenvalues lx and ly"""
    return lx*x, ly*y

# Grid of x, y points
nx, ny = 64, 64
x = np.linspace(-2, 2, nx)
y = np.linspace(-2, 2, ny)
X, Y = np.meshgrid(x, y)

# Define the plots
fig, ax = plt.subplots(1,3, figsize=(12,4))
plots = ['sink','saddle','source']
colors = [plt.cm.cool,plt.cm.inferno,plt.cm.brg]
eigenvalues = [[-2,-2],[-2,2],[2,2]]

for i in xrange(3):
    plot = plots[0]
    Ex, Ey = np.zeros((ny, nx)), np.zeros((ny, nx))
    ex, ey = E(X, Y, eigenvalues[i][0],eigenvalues[i][1])
    Ex += ex
    Ey += ey
    ax[i].streamplot(x, y, Ex, Ey, color=color, linewidth=2, cmap=colors[i], density=1, arrowstyle='fancy', arrowsize=2)
    if plot == 'saddle':
        ax[i].plot((-1,1),(0,0), '--', color = 'k')
        ax[i].plot((0,0),(-1,1), '--', color = 'k')
    ax[i].set_xlim(-1.5,1.5)
    ax[i].set_ylim(-1.5,1.5)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_aspect('equal')
    ax[i].xaxis.label.set_visible(False)
    ax[i].yaxis.label.set_visible(False)
    ax[i].axis('off')
    ax[i].add_artist(Circle((0,0), 0.1, color = 'k'))


plt.savefig('hyperbolic_new.png',dpi=500, bbox_inches='tight')
plt.show()
```
